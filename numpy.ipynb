Creating Array in Numpy

```python
import numpy as np

arr = np.array([1,2,3,4])
print(type(arr))
```

```python
# Using Linspace
arr = np.linspace(1,10,1000).reshape((10,10,10))
print(arr)
```

```python
arr = np.arange(0,1000).reshape(10,10,10)
print(arr.size)
```

```python
one_like = np.ones_like(arr)
print(one_like)
```

```python
# Creating Zero Like Array
zero_like = np.zeros_like(arr)
print(zero_like)
```

```python
# Empyte Like Array
empty_like = np.empty_like(arr)
print(empty_like)
```

```python
# One Matrix of (3,3)
one_matrix = np.ones((3,3))
print(one_matrix)
```

```python
# Zero Matrix of (3,3)
zero_matrix = np.zeros((3,3))
print(zero_matrix)
```

```python
# Empty Matrix of (4,5)
empty_matrix = np.empty((4,5))
print(empty_matrix)
```

```python
# Shape
shape_matrix = np.shape((4,5))
print(arr.shape)
```

```python
# Dimension
dimension_matrix = arr.ndim((4,5))
print(arr.ndim)
```

```python
# Size
size_matrix = arr.size
print(arr.size)
```
